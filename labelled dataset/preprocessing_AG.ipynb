```python
from transformers import BertTokenizer, BertModel

# Load the BERT model
model = BertModel.from_pretrained('sagorsarker/bangla-bert-base')

# Tokenize your text
tokenizer = BertTokenizer.from_pretrained('sagorsarker/bangla-bert-base')
text = "খাইসি খেয়েছি "
tokens = tokenizer.tokenize(text)
print("tokens",tokens)
# Convert your tokenized text to input IDs
input_ids = tokenizer.encode(text, add_special_tokens=False)
print("input",input_ids)
```

```python
from transformers import BertTokenizer, BertModel
import torch
# Load the BERT model
model = BertModel.from_pretrained('bert-base-uncased')

# Tokenize your text
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
text = "This is an example sentence."
tokens = tokenizer.tokenize(text)
print("tokens",tokens)
# Convert your tokenized text to input IDs
input_ids = tokenizer.encode(text, add_special_tokens=True)
print("input ids",input_ids)
# Pass the input IDs to the BERT model and get the output tensor
outputs = model(torch.tensor([input_ids]))
# print("output",outputs)
# Extract the word embedding from the output tensor
# print("output shape",outputs)
word_embedding = outputs[0][0]
print(word_embedding.shape)
```
